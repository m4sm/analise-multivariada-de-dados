Descrição das colunas:
 
df.describe() calcula automaticamente a média, valor mínimo e máximo das colunas numéricas(idade e salário)

```python
display(df.describe().round(2))
```

count: quantidade de linhas(pessoas cadastradas)
mean: média dos valores(28 é a idade média)
std: variação(as idades variam cerca de 5 anos em relação a média)
min: menor valor (a pessoa mais nova tem 22)
max: maior valor (a pessoa mais velha tem 35)
50%: valor que está no meio da lista (metade das pessoas tem menos de 28, metade tem mais)
