Utilizando sklearn.datasets para importar o dataset "Breast Cancer Wisconsin (Diagnostic)

```python
import pandas as pd
from sklearn.datasets import load_breast_cancer
```

Carregando os dados brutos do Dataset

```python
dados_cancer = load_breast_cancer()
```

Transformando o dataset em um DataFrame formatado e adicionando a coluna de alvo(prever se é maligno ou benigno)

```python
df = pd.DataFrame(dados_cancer.data, columns=dados_cancer.feature_names)
df['target'] = dados_cancer.target
```

Mostrando o tipo de cada coluna e verificando se há valores nulos

```python
df.info()
```
